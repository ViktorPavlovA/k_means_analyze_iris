#1) Preparing data

```python
from sklearn import datasets
import prettytable as prt
import numpy as np
```

```python
iris = datasets.load_iris()
print(iris)
```

```python
X = iris["data"]
y = iris["target"]

X =np.asanyarray(X)
y = np.asanyarray(y)
name_type_iris = iris["target_names"]
names_fl = []
for i in range(len(y)):
  if y[i] == 0:
    names_fl.append(name_type_iris[0])
  elif y[i] == 1:
    names_fl.append(name_type_iris[1])
  else: names_fl.append(name_type_iris[2])
print(X.shape)
print(y)
print(names_fl)
```

#2)Analyze stage with using K-means

```python
import pandas as pd
import matplotlib 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train,y_train)
```

```python
def metrics(y_pred,y_test):
  print(f"Scope: {np.mean(y_pred == y_test)}")
```

```python
y_pred = knn.predict(X_test)
metrics(y_pred,y_test)
```
